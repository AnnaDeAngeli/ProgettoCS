### Domanda 1:
#### Per ogni film, calcola il numero dei membri del cast

```python
#print delle prime righe della colonna cast:
credits.cast.head()
```

```python
#Imposto la funzione extract_dic che mi permette di estrarre i dizionari dalla colonna e creare un dataframe in cui si aggiunge
## l'ID del film in modo tale da poter ricondurre i membri del cast al film corrispondente.
def extract_dic(df, column):
    tables = df[column].apply(pd.read_json) #estraggo i dizionari dalla colonna e li trasformo in un dataframe (tables è una serie di dataframe)
    movie_id = df['movie_id']             #creo la serie movie_id uguale alla colonna movie_id del dataframe iniziale
    for i in range(len(df)):
        tables[i]['movie_id'] = df.movie_id[i]   #aggiungo colonna movie id in ogni dataframe in table
    return(pd.concat(tables.values).reset_index(drop = True))  #concateno 
```

```python
#applico la funzione extract_dic alla colonna cast nel dataframe credits
cast = extract_dic(credits, 'cast')
credits = credits.drop('cast', axis = 1) #elimino cast da credits
cast.head()
```

```python
#rinomino la colonna id con actor id per problemi di comprensibilità
cast = cast.rename(columns = {'id' : 'actor_id'})
```

```python
#calcolo il numero di membri del cast per ogni film
cast_members = cast.groupby(by = 'movie_id').count()
cast_members.rename(columns = {'cast_id' : '#cast_members'}, inplace = True)
cast_members = cast_members[['#cast_members']]
```

```python
#aggiungo il titolo del film al risultato
movie_tit = movies[['movie_id', 'title']]
number_cast_members = cast_members.merge(movie_tit, how = 'left', on = ['movie_id'])
number_cast_members = number_cast_members[['movie_id', 'title', '#cast_members']]
number_cast_members
```

### Domanda 4
#### Si estragga il dominio per ogni pagina

```python
#Si considera la colonna homepage della tabella movies
movies.homepage.head()
```

```python
homepages = movies[['movie_id', 'title', 'homepage']].dropna().reset_index(drop = True)
homepages.head()
```

```python
#Definisco una funzione per l'estrazione del dominio
def extract_domain(homepage):
    domains = []
    homepage = homepage.split(" ")
    for i in range(len(homepage)):
        domains.append(homepage[i].split('/')[2].split('.')[-1])
    return str(domains)[1:-1]
```

```python
#estraggo il dominio
homepages['domain'] = homepages.homepage.apply(extract_domain)
homepages
```

```python
#controllo i valori del dominio
homepages.domain.unique()
```

```python
#Il film Cargo è l'unico ad avere due domini
homepages[homepages.domain == "'ch', 'com'"][['title', 'homepage', 'domain']]
```

### Domanda 10
#### Per ogni company ed ogni decade, calcola i ricavi totali

```python
#E' necessario unire i dataframe movies e production_companies
m = movies[['year', 'revenue', 'movie_id']]
production_companies.head()
```

```python
m_prod = m.merge(production_companies, how = 'right', on = ['movie_id'])
m_prod.head()
```

```python
m_prod.year = m_prod.year.apply(int)
```

```python
#Definisco la funzione per l'estrazione della decade
def extract_decade(year):
    return(str(((year // 10)*10)) + '-' + str((((year // 10)*10)+10)))
```

```python
m_prod['decade'] = m_prod.year.apply(extract_decade) 
m_prod.head()
```

```python
m_prod_ = m_prod[['revenue', 'production_companies_id', 'name', 'decade']]
rev_per_decade_per_company = m_prod_.groupby(['production_companies_id', 'name', 'decade'],as_index = False).sum()
rev_per_decade_per_company.rename(columns = {'revenue' : 'overall_revenue'}, inplace = True)
rev_per_decade_per_company
```

### Domanda 11
#### Per ogni decade calcola la company con il ricavo massimo

```python
company_max_revenue_per_decade = m_prod.groupby('decade', as_index = False)[['revenue', 'name']].max()
company_max_revenue_per_decade
```

```python
for i in range(0, len(company_max_revenue_per_decade)):
    print(m_prod[m_prod["decade"]==company_max_revenue_per_decade.decade[i]][m_prod["revenue"]==company_max_revenue_per_decade.revenue[i]],"\n")
```
